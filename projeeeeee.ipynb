```python
import matplotlib.pyplot as plt
def input_graph():
    num_points = int(input("Grafik için veri noktası sayısını girin: "))

    x_values = []
    y_values = []

    for i in range(num_points):
        x = input(f"{i+1}. X değeri: ")
        y = float(input(f"{i+1}. Y değeri: "))
        x_values.append(x)
        y_values.append(y)

    return x_values, y_values

def create_graph():
    labels, data = input_graph()

    print("\nGrafik türlerini seçiniz:")
    print("1. Çizgi Grafiği\n2. Çubuk Grafiği\n3. Pasta Grafiği\n4. Saçılım Grafiği\n5. Histogram")
    graph_choice = int(input("Seçiminiz (1-5): "))

    if graph_choice == 1:
        line_graph(data, labels)
    elif graph_choice == 2:
        bar_graph(data, labels)
    elif graph_choice == 3:
        pie_chart(data, labels)
    elif graph_choice == 4:
        scatter_plot(data, labels)
    elif graph_choice == 5:
        histogram(data, labels)
    else:
        print("Geçersiz seçenek.")

def main():
    işlemler = [
        'Birim dönüştürme','Grafik oluşturma','Analiz hesaplayıcı'
    ]

    print("Yapmak istediğiniz işlemi seçiniz:")
    for işlem in işlemler:
        print(işlem)
    choice = input("Seçiminiz: ")

    if choice == 'Birim dönüştürme':
        value = float(input("Dönüştürmek istediğiniz değeri girin: "))
        converted_value = convert_units(value)
        print(f"Dönüştürülmüş değer: {converted_value} mg/L")

    elif choice == 'Grafik oluşturma':
        create_graph()

    elif choice == 'Analiz hesaplayıcı':
        analiz_program()
    else:
        print("Geçersiz seçenek. Lütfen listelenen işlemlerden birini seçin.")

def line_graph(data, labels):
    plt.plot(labels, data)
    plt.xlabel('Analiz Türü')
    plt.ylabel('Sonuçlar')
    plt.title('Çizgi Grafiği')
    plt.show()

def bar_graph(data, labels):
    plt.bar(labels, data)
    plt.xlabel('Analiz Türü')
    plt.ylabel('Sonuçlar')
    plt.title('Çubuk Grafiği')
    plt.show()

def pie_chart(data, labels):
    plt.pie(data, labels=labels, autopct='%1.1f%%')
    plt.title('Pasta Grafiği')
    plt.show()

def scatter_plot(data, labels):
    plt.scatter(labels, data)
    plt.xlabel('Analiz Türü')
    plt.ylabel('Sonuçlar')
    plt.title('Saçılım Grafiği')
    plt.show()

def histogram(data, labels):
    plt.hist(data, bins = len(labels))
    plt.xlabel('Sonuçlar')
    plt.ylabel('Frekans')
    plt.title('Histogram')
    plt.show()

def perform_analysis(choice):
     if choice == 'Toplam Sertlik':
        sarfiyat_degeri = float(input("EDTA sarfiyat degeri kactir? "))
        numune_hacmi = float(input("Numune hacmi değeri giriniz:"))
        result = (sarfiyat_degeri * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Alkalinite':
        asit_hacmi = float(input("Asit sarfiyatını giriniz:"))
        asit_normalitesi = float(input("Asit normalitesini giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = (asit_hacmi * asit_normalitesi * 50 * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Asidite':
        baz_hacmi = float(input("Baz sarfiyatını giriniz:"))
        baz_normalitesi = float(input("Baz normalitesini giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = (baz_hacmi * baz_normalitesi * 50 * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Yağ ve Gres Parametresi Tayini':
        a = float(input("Cam destilasyon balonunun son tartımını giriniz:"))
        b = float(input("Cam destilasyon balonunun darasını giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = ((a - b) * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Asidimetrik Metod ile Ammonyak Tayini (Su numunesi için)':
        a = float(input("Numune için sarf edilen H2SO4 sarfıyatını giriniz:"))
        b = float(input("Şahitini titrasyona için sarf edilen H2SO4 miktarını giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = ((a - b) * 280) / numune_hacmi
        print(result)
        return result
     if choice == 'Asidimetrik Metod ile Ammonyak Tayini (Çamur ve Sediment numunesi için)':
        a = float(input("Numune için sarf edilen H2SO4 sarfıyatını giriniz:"))
        b = float(input("Şahitini titrasyona için sarf edilen H2SO4 miktarını giriniz:"))
        c = float(input("Gram kuru numune ağırlığını giriniz:"))
        result = ((a - b) * 280) / c
        print(result)
        return result
     if choice == 'Askıda Katı Madde Tayini':
        a = float(input("Filtre darası giriniz:"))
        y = float(input("Kalıntılı filtre tartımı giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = ((y - a) * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Uçucu Askıda Katı Madde Tayini':
        a = float(input("103-105 C deki kurutma sonucu filtre tartımı giriniz:"))
        y = float(input("550-600 C yakma sonucu filtre tartımı giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = ((a - y) * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'Toplam Çönmüş Madde Tayini':
        a = float(input("Kapsul darası giriniz:"))
        y = float(input("Kalıntılı kapsul tartımı giriniz:"))
        numune_hacmi = float(input("Numune hacmini giriniz:"))
        result = ((y - a) * 1000) / numune_hacmi
        print(result)
        return result
     if choice == 'KOI':
       oxygen = float(input("Tüketilen oksijen miktarını giriniz:"))
       faktor = float(input("Seçilme faktörünü giriniz: "))
       numune_hacmi= float(input("Numune hacmini giriniz: "))
       result = oxygen*faktor/numune_hacmi
       print(result)
       return result
     if choice == 'Demir Tayını (Fe+3 konsantrasyonu)':
       numune_hacmi = float(input("Numune hacmini giriniz: "))
       edta_norm = float(input("EDTA normalitesi giriniz: "))
       edta_hacmi = float(input("EDTA sarfıyatı giriniz: "))
       result = ((edta_norm*edta_hacmi)*55.58*1000)/numune_hacmi
       print(result)
       return result
     if choice == 'BOI':
       oxygen = float(input("Başlangıçtaki çözünmüş oksijen seviyesini girin (mg/L): "))
       bes_gun = float(input("5 gün sonundaki çözünmüş oksijen seviyesini girin (mg/L): "))
       faktor = float(input("Seçilme faktörünü girin: "))
       result = (oxygen-bes_gun)*faktor
       print(result)
       return result
     if choice == 'Mangan (Mn+2 konsantrasyonu)':
       numune_hacmi = float(input("Numune hacmini giriniz: "))
       edta_norm = float(input("EDTA normalitesi giriniz: "))
       edta_hacmi = float(input("EDTA sarfıyatı giriniz: "))
       result = ((edta_norm*edta_hacmi)*54.94*1000)/numune_hacmi
       print(result)
       return result
     if choice == 'Toplam Katı Madde Tayini':
       a = float(input("Kapsul kalıntısını giriniz:"))
       b = float(input("Kapsül darasını giriniz: "))
       numune_hacmi = float(input("Numune hacmini giriniz: "))
       result = (a-b+b)*1000/numune_hacmi
       print(result)
       return result
     if choice == 'Klorür Tayini(mg/L Cl-)':
       a = float(input("Numune için sarfıyatnı giriniz:"))
       b = float(input("Şahit numune için sarfıyatını giriniz: "))
       numune_hacmi = float(input("Numune hacmini giriniz: "))
       norm = float(input("AgNO3 çözeltisinin normalitesi giriniz: "))
       result = (a-b)*norm*1000*35.45/numune_hacmi
       print(result)
       return result

def ask_for_unit_conversion():
    converted = input("Birim dönüştürme yapmak ister misiniz? (evet/hayır): ")
    return converted.lower() in ['evet', 'e']

def convert_units(value):
    print("Birim Dönüştürücü")
    print("1. ton/L'den mg/L'e")
    print("2. ton/m^3'den mg/L'e")
    print("3. kg/L'den mg/L'e")
    print("4. kg/m^3'den mg/L'e")
    print("5. g/L'den mg/L'e")
    print("6. g/m^3'den mg/L'e")

    choice = int(input("Dönüştürme seçeneğinizi giriniz (1-6): "))

    if choice == 1:
      result = value * 1e6
    elif choice == 2:
      result = value * 1e3
    elif choice == 3:
      result = value * 1e6
    elif choice == 4:
      result = value * 1e3
    elif choice == 5:
      result = value * 1e3
    elif choice == 6:
      result = value
    else:
        print("Geçersiz seçenek.")
        return None

    return result

def analiz_program():
    analyses = [
        'KOI', 'BOI', 'Toplam Sertlik (CaCO3 mg/L)', 'Alkalinite (CaCO3 mg/L)', 'Asidite (CaCO3 mg/L)',
        'Yağ ve Gres Parametresi Tayini',
        'Asidimetrik Metod ile Ammonyak Tayini (Su numunesi için)',
        'Asidimetrik Metod ile Ammonyak Tayini (Çamur ve Sediment numunesi için)',
        'Askıda Katı Madde Tayini', 'Uçucu Askıda Katı Madde Tayini',
        'Toplam Çönmüş Madde Tayini','Demir Tayını (Fe+3 konsantrasyonu)','Mangan (Mn+2 konsantrasyonu)',
        'Toplam Katı Madde Tayini'
    ]

    print("Çevrimiçi Analiz Hesaplayıcına Hoş Geldiniz!")
    print("Lütfen yapmak istediğiniz işlemi seçiniz:")

    for analysis in analyses:
        print(analysis)

    choice = input("Analiz seçimi (örn., KOI): ")

    create_graph = input("Bir grafik oluşturmak ister misiniz? (evet/hayır): ").lower()
    if create_graph == 'evet':
        results = []
        labels = [f'{choice} {i+1}' for i in range(5)]
        for i in range(5):
            result = perform_analysis(choice)
            if result is not None:
              results.append(result)
              if ask_for_unit_conversion():
                 converted_result = convert_units(result)
                 print("Dönüştürülen değer:", converted_result, "mg/L")
            else:
                print("Analiz sonucu geçersiz.")

        print("\nGrafik türlerini seçiniz:")
        print("1. Çizgi Grafiği\n2. Çubuk Grafiği\n3. Pasta Grafiği\n4. Saçılım Grafiği\n5. Histogram")
        graph_choice = int(input("Seçiminiz (1-5): "))

        if graph_choice == 1:
            line_graph(results, labels)
        elif graph_choice == 2:
            bar_graph(results, labels)
        elif graph_choice == 3:
            pie_chart(results, labels)
        elif graph_choice == 4:
            scatter_plot(results, labels)
        elif graph_choice == 5:
            histogram(results, labels)
        else:
            print("Geçersiz seçenek.")
    else:
        result = perform_analysis(choice)
        if result is not None:
            print("Analizin sonucu:", result)
            if ask_for_unit_conversion():
                converted_result = convert_units(result)
                print("Dönüştürülen değer:", converted_result, "mg/L")
        else:
            print("Analiz sonucu geçersiz.")

if __name__ == "__main__":
    main()
```
